```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
import random
from tqdm import tqdm
import os
import pickle
from joblib import Parallel, delayed

from utils import *
```

```python
output_dir = "data"
os.makedirs("data", exist_ok=True)
#num_trials = 30
num_trials = 5

# fixed variables
dimension = 32
context_size = 8
matrix_normalization_method = "fro"
decay = True
rate = 0.85
dataset_size = int(1e6)
```

```python
# Variance Sweep
variances = [1e-3, 1e-2, 1e-1, 1]
```

```python
def generate_trial_data(variance, trial, dimension, context_size, dataset_size, 
                       matrix_normalization_method, decay, rate, data_dir):
    """Generate data for a single trial if pickle doesn't exist."""
    variance_dir = os.path.join(data_dir, str(variance))
    filename = f"trial_{trial}.pkl"
    filepath = os.path.join(variance_dir, filename)
    
    # Check if pickle file already exists
    if os.path.exists(filepath):
        print(f"Skipping Variance: {variance}, Trial: {trial} (file exists)")
        return None
    
    print(f"Generating Variance: {variance}, Trial: {trial}")
    #reset_seeds(trial)
    data = generate_random_data(dimension, context_size, dataset_size, variance,
                               matrix_normalization_method, decay, rate)
    
    # Pickle the data tuple
    with open(filepath, 'wb') as f:
        pickle.dump(data, f)
    
    return f"Completed Variance: {variance}, Trial: {trial}"

# Main execution
data_dir = os.path.join(output_dir)
os.makedirs(data_dir, exist_ok=True)

# Create variance directories
for variance in variances:
    variance_dir = os.path.join(data_dir, str(variance))
    os.makedirs(variance_dir, exist_ok=True)

# Create all (variance, trial) combinations
tasks = [(variance, trial) for variance in variances for trial in range(num_trials)]

# Run in parallel increase n_jobs as needed
results = Parallel(n_jobs=5)(
    delayed(generate_trial_data)(
        variance, trial, dimension, context_size, dataset_size,
        matrix_normalization_method, decay, rate, data_dir
    ) for variance, trial in tasks
)

print("All tasks completed!")
```
